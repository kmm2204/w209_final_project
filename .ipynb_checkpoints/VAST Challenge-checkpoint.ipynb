```python
df = pd.read_csv("/Users/robertdeng/Google Drive/Data Science/W209/Project/Lekagul Sensor Data.csv")
```

```python
#1) Plot a histogram of car types at entrance 1.
entrance1 = sorted(list(df.loc[lambda df : df["gate-name"] == "entrance1", "car-type"]))

n, bins, patches = plt.hist(entrance1, bins = range(0,7), facecolor='g', alpha=0.75)

plt.xlabel('Car Type')
plt.ylabel('Car Counts')
plt.title('Histogram of Car Types at Entrance 1')
plt.axis([0, 6, 0, 3000])
plt.grid(True)
plt.show()
```

```python
#2) Plot histograms of car types at all entrances
all_entrances = sorted(list(df["car-type"]))

n, bins, patches = plt.hist(all_entrances, facecolor='red', alpha=0.75)

plt.xlabel('Car Type')
plt.ylabel('Frequency')
plt.title('Histogram of Car Types at All Entrances')
plt.axis([0, 6, 0, 70000])
plt.grid(True)
plt.show()
```

```python
#3) Plot the count of daily visits grouped by day for the whole dataset.
g = df.groupby(pd.DatetimeIndex(pd.to_datetime(df["Timestamp"])).date).count()
plt.figure(figsize=(15, 5))
plt.plot(g,"y-", alpha = 0.5)
plt.show()
```

```python
#4) Same as plot (3), but plot a separate line for each car type.
plt.figure(figsize=(20, 10))

for car_type in df["car-type"].unique(): 
    g = df[df["car-type"] == car_type].groupby(pd.DatetimeIndex(df[df["car-type"] == car_type]["Timestamp"]).date).count()
    plt.plot(g, "-", alpha = 0.5)
    
plt.xlabel('Date (Days)')
plt.ylabel('Daily Visits')
plt.title('Daily Visits by Car Type')
plt.legend(df["car-type"].unique())
plt.show()
```

```python
#5) Repeat plot 3, but group by the hour.
plt.xlabel('Daily Visits By Hour')
plt.ylabel('Daily Visits')
plt.title('Hour of Day')
plt.plot(df[df["car-type"] == "4"].groupby(pd.DatetimeIndex(df[df["car-type"] == "4"]["Timestamp"]).hour).count().iloc[:,1], "c--")
plt.show
```

```python
#6) Repeat plot 4, grouping by the day of week, and show a stacked bar chart for each car type.

#a. Hint 1: pd.DatetimeIndex has the attribute .dayofweek.

#b. Hint 2: You can loop over the car types again, and to stack the bars, use plt.bar() with argument bottom=prev 
#(where prev is the height of the bars from before). 
#It may help to start with prev = np.zeros(7) outside of the loop, 
#and add to it at the end of each loop (prev += g.count().Timestamp.values)

prev = np.zeros(7)
for car_type in sorted(df["car-type"].unique()):
    g = df[df["car-type"] == car_type].groupby(pd.DatetimeIndex(df[df["car-type"] == car_type]["Timestamp"]).dayofweek)
    plt.bar(g.count().index.values, g.count().Timestamp, bottom = prev)
    prev += g.count().Timestamp.values
    


plt.title('Car Type Visits by Day of Week')
plt.legend(df["car-type"].unique(), loc='center left', bbox_to_anchor=(1, 0.5))
dow_labels = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
plt.xlabel('Day of Week')
plt.xticks(range(7), dow_labels, rotation='vertical')
plt.ylabel('Visits')
plt.show()
```
